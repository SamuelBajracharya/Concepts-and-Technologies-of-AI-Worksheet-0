```python
# TASK 1

def min_coins(coins, amount):
    """
    This function solves the coin change problem by determining the fewest number of
    coins from a given set of coin denominations that sum up to a target amount. The
    solution uses dynamic programming(tabulation) to iteratively build up the minimum
    number of coins required for each amount.

    Returns:
    int: The minimum number of coins required to make the given amount.
    If it is not possible to make the amount with the given coins, returns -1.
    Example:
    >>> min_coins([1, 2, 5], 11)
    3
    >>> min_coins([2], 3)
    -1
    """
    dp = [float('inf')] * (amount + 1)
    dp[0] = 0
    for coin in coins:
        for i in range(coin, amount + 1):
            dp[i] = min(dp[i], dp[i - coin] + 1)
    return dp[amount] if dp[amount] != float('inf') else -1

coins = [1,2,5]
amount = 11

print(min_coins(coins, amount))

# TASK 2

def longest_common_subsequence(s1, s2):
    """
    This function takes two strings as input and finds the length of their longest
    common subsequence.
    Args:
    s1(string): first string
    s2(string):second string
    Returns:
    int: the length of the longest common subsequence of the provided strings
    """
    m, n = len(s1), len(s2)
    dp = [[0] * (n + 1) for _ in range(m + 1)] #cretes a nested list of length of m including lists of 0 with length of n

    for i in range(1, m + 1): #iterates through the outermost list
        for j in range(1, n + 1): #iterates through the innermost list
            if s1[i - 1] == s2[j - 1]: #checks if the character is the of the alphabet is the same
                dp[i][j] = dp[i - 1][j - 1] + 1
            else:  # if characters don't match
                dp[i][j] = max(dp[i - 1][j], dp[i][j - 1])
    return dp[m][n]

s1 = "ABCDE"
s2 = "ACE"
length_lcs = longest_common_subsequence(s1, s2)
print(f"The length of the LCS of '{s1}' and '{s2}' is: {length_lcs}")


# TASK 3

def knapsack(weights, values, capacity):
    """
        This function takes in a list of weights, a list of values, and a maximum capacity for
        the knapsack, and returns the maximum value that can be obtained while keeping
        the total weight within the given capacity.
        Args:
        weights(list of int): a list of weights of items
        values(list of int): a list of values of items
        capacity(int): the maximum capacity of the knapsack
        Returns:
        int: the maximum value that can be obtained
        """

    w_len = len(weights) #length of weights
    dp = [[0] * (capacity + 1) for _ in range(w_len + 1)] #creating a 2D list filled with innitial value of 0

    for w in range(1, w_len + 1): # looping over the outermost list
        for i in range(capacity + 1): # looping over the innermost list
            if weights[w - 1] <= i:
                dp[w][i] = max(dp[w - 1][i], values[w - 1] + dp[w - 1][i - weights[w - 1]]) #replaces the value of dp[w][i] with max value
                # print(dp)
            else:
                dp[w][i] = dp[w - 1][i]

    return dp[w_len][capacity]


weights = [1, 3, 4, 5]
values = [1, 4, 5, 7]
capacity = 7
result = knapsack(weights, values, capacity)
print(f"Maximum value that can be obtained with {capacity} capacity is: {result}")
```
